```python
#txt파일 읽어와서 class id 수정해서 저장 후, label의 경로를 label.txt에 작성
```

```python
f = open("/home/petpeotalk/Desktop/park/ScaledYOLOv4/labeling/label.txt", 'a') # a로
change_value_dict = {0: 16, 1: 15, 2: 0}

for data in train_datalist:
    data_path = os.path.join(train_data_path, data)
    df = pd.read_csv(data_path, sep=" ",header=None) # data_path에 해당하는 파일을 data frame으로 읽어옴
    #df_target = df[(df[0]==0) | (df[0]==15) | (df[0]==16)] # 0번째 행, 즉 class가 0 ,15, 16 각각 사람 고양이 개에 해당하는 것만 걸러냄
    
    df = df.replace ({0: change_value_dict})
    
    result_path = os.path.join(result,data) # 저장할 경로 지정
    df.to_csv(result_path, sep = ' ', index = False, header=None) #txt파일로 저장.
        
    img = data.replace("txt","jpg")
    img_path = os.path.join(train_img_path,img)
    new_img_path = img_path + '\n'
    f.write(new_img_path)

f.close()
```
